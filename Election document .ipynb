```python
import pandas as pd

# Load the data with a specified encoding
df = pd.read_csv('C:/Users/Lenovo/Documents/Elections Result.csv', encoding='ISO-8859-1')

# Convert 'Votes' column to numeric
df['Won'] = pd.to_numeric(df['Won'], errors='coerce')

# Drop rows with missing values
df.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt

# Total votes by party
party_votes = df.groupby('Party')['Won'].sum().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(40, 20))
bars = party_votes.plot(kind='bar', color='orange')
plt.title('Total Won by Party', fontsize=30)
plt.xlabel('Party', fontsize=20)
plt.ylabel('Total', fontsize=20)
plt.xticks(rotation=90, fontsize=20)
plt.yticks(fontsize=16)

# Adding annotations (numbers) on top of the bars
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, int(yval), ha='center', va='bottom', fontsize=16)

plt.tight_layout()
plt.show()
```

```python
# Voter turnout by constituency
constituency_turnout = df.groupby('Party')['Won'].sum().sort_values(ascending=False).head(10)

# Plotting
plt.figure(figsize=(40, 20))
constituency_turnout.plot(kind='bar', color='green')
plt.title('Top 10 Parties with Highest Voter Turnout', fontsize=30)
plt.xlabel('Constituency', fontsize=20)
plt.ylabel('Total Won', fontsize=20)
plt.xticks(rotation=60, fontsize=20)
plt.tight_layout()
plt.show()
```
